```python
# app/app.py

import streamlit as st
import pandas as pd
from collections import Counter

# Cargar datos preparados
@st.cache_data
def load_data():
    df = pd.read_csv("C:/Users/druiz/Documents/Proyectos data analyst/Proyecto Cocktails/data/merged_cocktails.csv", converters={"ingredient": eval})
    return df

df = load_data()

st.title("üçπ Cocktail Selector")

st.markdown("### Paso 1: Elige un ingrediente para comenzar")
st.markdown("Estos son los ingredientes m√°s utilizados en el dataset:")

# Obtener los m√°s comunes
ingredientes_planos = [ing.lower().strip() for lista in df["Ingredient"] for ing in lista]
top_ingredientes = [i[0] for i in Counter(ingredientes_planos).most_common(20)]

# Carrusel horizontal simulado
cols = st.columns(len(top_ingredientes))
seleccionado = None
for i, col in enumerate(cols):
    if col.button(top_ingredientes[i].capitalize()):
        seleccionado = top_ingredientes[i]

# Mostrar selecci√≥n
if seleccionado:
    st.markdown("---")
    st.subheader(f"üîç Has elegido: {seleccionado.capitalize()}")
    st.info("En el siguiente paso ver√°s ingredientes que suelen aparecer con este.")

    # Aqu√≠ podr√≠as guardar `seleccionado` en `st.session_state` para continuar con Paso 2
```
