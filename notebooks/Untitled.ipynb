```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aplanamos la lista de ingredientes
from collections import Counter

todos_los_ingredientes = [ing.strip().lower() for sublist in df_merged['Ingredient'] if isinstance(sublist, list) for ing in sublist]
conteo_ingredientes = Counter(todos_los_ingredientes)
ingredientes_comunes = conteo_ingredientes.most_common(20)

# Lo convertimos a DataFrame para graficar
df_freq = pd.DataFrame(ingredientes_comunes, columns=['Ingrediente', 'Frecuencia'])

# Gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(data=df_freq, y='Ingrediente', x='Frecuencia', palette='viridis')
plt.title('🥃 Ingredientes más frecuentes en cócteles')
plt.xlabel('Frecuencia')
plt.ylabel('Ingrediente')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np

# Asegurarse de que todas las listas estén bien formateadas
ingredientes_filtrados = df_merged[df_merged['Ingredient'].notnull()]['Ingredient']

# Binarizamos los ingredientes por cóctel (one-hot)
mlb = MultiLabelBinarizer()
ingredientes_binarios = mlb.fit_transform(ingredientes_filtrados)

# Creamos la matriz de coocurrencia
coocurrencia = np.dot(ingredientes_binarios.T, ingredientes_binarios)
coocurrencia_df = pd.DataFrame(coocurrencia, index=mlb.classes_, columns=mlb.classes_)
```

```python
# Tomamos los 20 ingredientes más comunes
ingredientes_top = coocurrencia_df.sum(axis=1).sort_values(ascending=False).head(20).index.tolist()

coocurrencia_top = coocurrencia_df.loc[ingredientes_top, ingredientes_top]

# Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(coocurrencia_top, cmap='YlOrBr', annot=True, fmt='d', linewidths=.5)
plt.title('🔥 Coocurrencia entre los 20 ingredientes más frecuentes')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
X = mlb.fit_transform(df_merged['Ingredient'].dropna())

ingredientes_df = pd.DataFrame(X, columns=mlb.classes_)
ingredientes_df.index = df_merged[df_merged['Ingredient'].notnull()].index  # Alineamos con el DataFrame original
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=42)
clusters = kmeans.fit_predict(ingredientes_df)

# Añadimos el resultado al DataFrame original
df_merged.loc[ingredientes_df.index, 'cluster'] = clusters
```

```python
df_merged['cluster'].value_counts()
```

```python
for i in range(6):
    print(f"\n--- Clúster {i} ---")
    subgrupo = ingredientes_df[df_merged['cluster'] == i]
    top = subgrupo.sum().sort_values(ascending=False).head(10)
    print(top)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Usamos la misma matriz binaria de ingredientes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(ingredientes_df)

# Creamos un DataFrame con los resultados
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['cluster'] = df_merged.loc[ingredientes_df.index, 'cluster'].astype(int)

# Visualización
plt.figure(figsize=(10,7))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='cluster', palette='Set2', s=60)
plt.title('🔍 Clustering de cócteles basado en ingredientes (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Clúster')
plt.tight_layout()
plt.savefig('PCA.png')
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(ingredientes_df)

df_tsne = pd.DataFrame(X_tsne, columns=['TSNE1', 'TSNE2'])
df_tsne['cluster'] = df_merged.loc[ingredientes_df.index, 'cluster'].astype(int)

plt.figure(figsize=(10,7))
sns.scatterplot(data=df_tsne, x='TSNE1', y='TSNE2', hue='cluster', palette='Set2', s=60)
plt.title('🌀 Clustering de cócteles con t-SNE')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(title='Clúster')
plt.tight_layout()
plt.savefig('t-sne.png')
plt.show()

```

```python
def buscar_cocteles_por_ingredientes(df, ingredientes_usuario, tolerancia=1.0):
    """
    Filtra cócteles según los ingredientes disponibles del usuario.

    Args:
        df (DataFrame): DataFrame que debe contener una columna 'ingredient' con listas de ingredientes.
        ingredientes_usuario (list): Lista de ingredientes disponibles.
        tolerancia (float): Fracción mínima de ingredientes del cóctel que deben estar presentes.
                            1.0 = coincidencia exacta, 0.6 = al menos 60%.

    Returns:
        DataFrame filtrado con los cócteles que cumplen el criterio.
    """
    def calcular_coincidencia(fila):
        ingredientes_receta = set(fila)
        ingredientes_disponibles = set(ingredientes_usuario)
        interseccion = ingredientes_receta & ingredientes_disponibles
        return len(interseccion) / len(ingredientes_receta) if ingredientes_receta else 0

    df_filtrado = df.copy()
    df_filtrado['porcentaje_coincidencia'] = df_filtrado['ingredient'].apply(calcular_coincidencia)
    return df_filtrado[df_filtrado['porcentaje_coincidencia'] >= tolerancia].sort_values(
        by='porcentaje_coincidencia', ascending=False
    )
```
